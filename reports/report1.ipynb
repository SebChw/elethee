## ELECTRE TRI-B

````python
get_df = lambda ar: pd.DataFrame(ar, columns = data.columns, index= ["B1", "B2", "B3"])
````

````python
attributes_types = ["gain"] * 5 + ["cost"] * 2
pd.DataFrame(np.array([attributes_types]), columns=data.columns)
````

First profiles and threshold values were somehow inferred based on inspection of column values. It was done rather intuitivelly.

````python
profiles = np.array(
    [
        [5, 60, 60, 30, 30, 400, 1],
        [30, 80, 80, 50, 60, 250, 0.8],
        [60,90,90,60, 70,150,0.7]
    ]
)

get_df(profiles)
````

````python
thresholds = {
    "q": np.array(
        [
            [2, 5, 5, 5, 5, 20, 0.1],
            [10, 5, 4, 4, 4, 20, 0.1],
            [10, 3, 2, 3, 3, 30, 0.05]
        ]),
    "p": np.array(
        [
            [6, 10, 7, 8, 6, 30, 0.15],
            [13, 8, 5, 7, 7, 30, 0.15],
            [11, 4, 3, 4, 5, 40, 0.1],
        ]),
    "v": np.array([
        [np.nan, 5, 5, 10, 10, np.nan, np.nan],
        [np.nan, 10, 14,10, 10, np.nan, np.nan],
        [np.nan, 10, 10, 10, 10, np.nan, np.nan]
    ])
}
````

````python
print("INDIFFERENCE THRESHOLDS")
get_df(thresholds['q'])
````

````python
print("PREFERENCE THRESHOLDS")
get_df(thresholds['p'])
````

````python
print("VETO THRESHOLDS")
get_df(thresholds['v'])
````

We select credibility threshold automatically with heuristic given on the lecture

````python
WEIGHTS_ARR = np.array([WEIGHTS[name] for name in data.columns])
CRED_TH = infer_lambda(WEIGHTS_ARR, [[2, 3], [1,2], [3,4]])
CRED_TH
````

````python
from mcda.electretrib import ElectreTriB
solver = ElectreTriB()

PREFERENCE_TYPES = np.array([1 if x=="gain" else -1 for x in attributes_types])
````

````python
def get_results(data, thresholds, profiles):
    pessimistic = solver.solve(data.to_numpy(), PREFERENCE_TYPES,
                   thresholds, profiles, WEIGHTS_ARR, CRED_TH, "pessimistic")

    optimistic = solver.solve(data.to_numpy(), PREFERENCE_TYPES,
                   thresholds, profiles, WEIGHTS_ARR, CRED_TH, "optimistic")
    data_with_assign = data.copy()
    data_with_assign["pessimistic"] = pessimistic
    data_with_assign["optimistic"] = optimistic

    return data_with_assign.sort_values("pessimistic")
````

A priori we gave assignments as such:
* Class 1 -> D
* Class 2 -> U
* Class 3 -> B
* Class 4 -> I

````python
get_results(data, thresholds, profiles)
````

Our DM is quite happy with the classification. However, he thinks that the profiles are too hard to beat and too many alternatives are in 1st class from the pessimistic perspective. He wants to change parameters slightly to make profiles weaker. Moreover it doesn't comply with a priori assignments

He decreases values of profile for:
* 1st attribute - WSK_ocena_p._kadre from [5,30,60] to [3,20,30] and 2nd attribute - ekonom losy absolwenta [60,80,90] to [55,65,80]. These attributes are not that important for decision maker so we probably can make them little bit smaller.
* Decrease attributes for `wspolprace zagraniczne for profile 2 and 3 [60,70] -> [50,65] as it was too strict
* decreased values for attribute [1,2,3] for the last profile by 5 [90,90,60] -> [85,85,55]
* Increased indifference and preference thresholds of attributes [1,2,3,4] by 3 and veto as veto by 10 as it was too small


````python
profiles[:,0] = [3,20,50]
profiles[:,1] = [55,65,80]
profiles[[1,2],4] = [50,65]
profiles[-1, [2,3]] -= 5
thresholds['q'][:, [1,2,3,4]] +=3
thresholds['p'][:, [1,2,3,4]] +=3
thresholds['v'][:, [1,2,3,4]] +=10
````

````python
get_results(data, thresholds, profiles)
````

He likes this ranking much more, he will be interesed in universities from group 3 to 4. 

Let's assume that he found a music band in his hometown se now he wants to come back much more often so distance is very important to him.

````python
Z = 10
ORDERED_CRITERIA = ["progi", "WSK_ocena_p._kadre", "ekonom._losy_absolw.", 
                    ("TOP_10", "pub._we_wspl._zagr."), "jak._przyj._na_studia", "distance"]
BLANK_CARDS = [0,1,0,1,3]

WEIGHTS = get_srf(Z, ORDERED_CRITERIA, BLANK_CARDS)
WEIGHTS_ARR = np.array([WEIGHTS[name] for name in data.columns])
CRED_TH = infer_lambda(WEIGHTS_ARR, [[2, 5], [1,5], [5,4]])
pd.DataFrame(WEIGHTS.values(), columns = ["weights"], index=WEIGHTS.keys())

thresholds['v'][:-1, :] = 100
````

````python
get_results(data, thresholds, profiles)
````

Now this changed the situation a lot. Now universities far away are much less preffered.
